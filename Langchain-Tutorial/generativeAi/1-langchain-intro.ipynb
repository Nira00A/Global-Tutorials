```python
from langchain.chat_models import init_chat_model
from dotenv import load_dotenv

load_dotenv() # Loads your GROQ_API_KEY from .env

# This provides the missing 'model' argument
model = init_chat_model("llama3-8b-8192", model_provider="groq")

# Now this will work!
response = model.invoke("Write me an essay on AI")
print(response.content)
```

```python
### Using CHATGROQ 

from langchain_groq import ChatGroq
model = ChatGroq(
    model = "llama-3.3-70b-versatile",
    temperature=0.0,
    max_retries=2
)

response = model.invoke("Write me an essay on AI. Must contain atmost 100 words")
print(response.content)
```

### Generative AI using Google Gemini

```python
from langchain_google_genai import ChatGoogleGenerativeAI

model = ChatGoogleGenerativeAI(
    model= 'gemini-2.5-flash',
    temperature= 0.0,
    max_retries= 2
)

response = model.invoke("Write me an essay on AI. Must contain atmost 100 words")
print(response.content)
```

### Streaming And Batch Processing

#### Streaming 

Most models can stream their output content while beign generated. This can be used to display the output to the user in real-time.

```python
for chunk in model.stream("Write ne an 1000 words paragraph on AI"):
    print(chunk.text , end="\n" , flush=True)
```

### Batch

Batching is a collection of independent requests to a model can significantly improve performance and reduce costs, as the processing can be done in parallel.

```python
responses = model.batch(
    [
        "What is AGI?",
        "What is AI?",
        "What is ML?"
    ]
)

for response in responses:
    print(response.text)
```
